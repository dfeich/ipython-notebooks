The forced Oscillator
=====================

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [12.0, 8.0]
```

```python
from scipy.integrate import odeint, ode
import numpy as np
import matplotlib.pyplot as plt
```

We consider a driving force $F_{driving} = F_0 cos(\omega_{dr} t)$. The differential equation for a forced oscillator with damping proportional to the velocity is then

$\displaystyle m \frac{\mathrm{d}^2x}{\mathrm{d}t^2} + m \gamma\frac{\mathrm{d}x}{\mathrm{d}t} + m \omega^2_0 x + F_0 cos(\omega_{dr} t) = 0$

To write this second-order ODE on standard form we introduce $p = m \frac{\mathrm{d}x}{\mathrm{d}t}$:

$\displaystyle \frac{\mathrm{d}p}{\mathrm{d}t} = - \gamma p - m \omega^2_0 x - F_0 cos(\omega_{dr} t)$

$\displaystyle \frac{\mathrm{d}x}{\mathrm{d}t} = \frac{p}{m}$


```python
def dy(y, t, gamma, w0, m, F0, wdr):
    """function takes x and p and returns their derivatives. The rest of the arguments
    are the parameters"""
    x, p = y[0], y[1]
    
    dx = p / m
    dp = - gamma * p - m * w0 * w0 * x -F0 * np.cos(wdr * t)
    
    return [dx, dp]
```

Resonance
---------

```python
# initial state
y0 = [0.0, 0.0]
```

```python
# time coodinate to solve the ODE for
t = np.linspace(0, 10, 1000)
w0 = 2*np.pi*1.0

m = 1.0
F0 = 1.0
wdr = w0
```

```python
y_a = odeint(dy, y0, t, args = (0.0, w0, m, F0, w0))
y_b = odeint(dy, y0, t, args = (0.0, w0, m, F0, 0.5*w0))
y_c = odeint(dy, y0, t, args = (0.0, w0, m, F0, 1.5*w0))
```

Plot showing the time evolution for the case of a Force acting with the resonance frequency $\omega_0$, and curves for forces acting at 0.5 and 1.5 times of $\omega_0$

```python
fig, ax = plt.subplots()
ax.plot(t, y_a[:,0],'k',linewidth=0.25, label="$w_{dr} = w_0$")
ax.plot(t, y_b[:,0], label="$w_{dr} = 0.5 w_0$")
ax.plot(t, y_c[:,0], label="$w_{dr} = 1.5 w_0$")
ax.legend()
plt.ylabel('x')
plt.xlabel('t')
```

```python
y_d = odeint(dy, y0, t, args = (0.0, w0, m, F0, 10*w0))
y_e = odeint(dy, y0, t, args = (0.0, w0, m, F0, 0.1*w0))

fig, ax = plt.subplots()
#ax.plot(t, y_a[:,0],'k',linewidth=0.25, label="$w_0 = w_{dr}$")
ax.plot(t, y_d[:,0], label="$w_{dr} = 10 w_0$")
ax.legend()
plt.ylabel('x')
plt.xlabel('t')

fig, ax = plt.subplots()
#ax.plot(t, y_a[:,0],'k',linewidth=0.25, label="$w_0 = w_{dr}$")
ax.plot(t, y_e[:,0], label="$w_{dr} = 0.1 w_0$")
ax.legend()
plt.ylabel('x')
plt.xlabel('t')
```
