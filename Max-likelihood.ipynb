===========

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import scipy.stats as stats
```

```python
plt.rcParams['figure.figsize'] = [12.0, 8.0]
```

A test function and a number of sample points
---------------------------------------------

```python
true_params = [0.1,1,5]
ftrue = np.poly1d(true_params)
```

```python
numsamples = 1000
(xmin,xmax) = (0,10)
```

We create the _measurements_ with some added random noise (thereby also building in that the errors follow a Gaussian distribution).

```python
#noise_max = 4
xax = np.linspace(xmin,xmax,100)
xi = np.random.rand(numsamples)*(xmax-xmin)
#yi = f(xi) + (rand(numsamples)-0.5)*2*noise_max
yi = ftrue(xi) + np.random.normal(loc=0.0,scale=2.0, size = numsamples)
plt.plot(xax,ftrue(xax),xi,yi,'.')
```

Max Likelihood
--------------

We choose a model function $f_{model}$ that we want to fit. We also need to choose a distribution function for our noise. Here we choose a *normal distribution*. We then assume that we can describe the distribution 

For an assumed set of parameters $p$, we can calculate the probability $f(x_i,y_i; p)$ of measuring a particular $x_i, y_i$. The likelihood of obtaining the whole series of measurements given this set of parameters is the product of all the probabilities and defines the maximal likelihood function.

$$ L = \prod_{i} f(x_i,y_i; p)$$
$$ f(x_i,y_i; p) = \frac{1}{\sigma \sqrt{2 \pi}} exp \left( - \frac{(y_i - f_{model}(x_i,y_i; p))^2}{2 \sigma^2} \right) $$

```python
def fmodel(x,params):
    return params[0]*x*x + params[1]*x + params[2]
```

The corresponding Likelihood function needs to be maximized. It is customary to maximize the logarithm of the function in order to turn the product over n into a sum. Since our optimization library offers minimizers, we use the negative of the logarithm of the likelihood function.

```python
def neg_log_likel(params, x, yobserved):
    modelparams = params[0:-1]
    sigma = params[-1]
     
    return -np.sum( stats.norm.logpdf(yobserved, loc=fmodel(x,modelparams), scale=sigma) )
```

We guess at the initial parameters and use the numerical optimization

```python
initparams = [0.1,1,3,0.3]
results = minimize(neg_log_likel, initparams, (xi,yi), method="Nelder-mead")
results
```

```python
modelparams = results.x[0:-1]
sigma = results.x[-1]
```

```python
fig1 = plt.figure()
plt.plot(xax,f(xax), label = "true function", )
plt.plot(xi,yi,'.', label = "samples")
plt.plot(xax, fmodel(xax, modelparams),'r-', label = "MLM fit")
plt.plot(xax, fmodel(xax, modelparams) + sigma,'r:',
     xax, fmodel(xax, modelparams) - sigma,'r:',label = "sigma")
plt.legend(loc='best')
```

```python
plt.hist(yi - fmodel(xi, modelparams), bins=21, normed=True)
plt.plot(np.linspace(-5,5,100), stats.norm.pdf(np.linspace(-5,5,100), loc=0, scale = 2.0, ))
```

The Chi Square value
-----

$\chi^2 = (\frac{y_1 - f(x_1)}{\sigma_1})^2 + (\frac{y_2 - f(x_2)}{\sigma_2})^2 + ... + (\frac{y_n - f(x_n)}{\sigma_n})^2$

The number of *effective* data points, i.e., those that could not be automatically hit by the curve, is called the number of *degrees of freedom*

degrees of freedom (dof) = number of data points - number of adjustable parameters 


```python
chisq = sum((yi - fmodel(xi, modelparams)) * (yi - fmodel(xi, modelparams)) /
            (sigma * sigma))
chisq
```

```python
dof = len(xi) - len(modelparams)
dof
```

The *reduced Chi Square value* is the $\chi^2$ value divided by the degrees of freedoms

```python
chisq/dof
```
