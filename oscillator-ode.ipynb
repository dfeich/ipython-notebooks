```python
from scipy.integrate import odeint, ode
import matplotlib.pyplot as plt
```

The differential equation for a damped oscillator with no external acting force is

$\displaystyle m \frac{\mathrm{d}^2x}{\mathrm{d}t^2} + m \gamma\frac{\mathrm{d}x}{\mathrm{d}t} + m \omega^2_0 x = 0$

The analytic solutions are of the nature

$x = e^{-\gamma t / 2} (A e^{i \omega_{\gamma} t} + A^* e^{i \omega_{\gamma} t})$ ;    with $\omega_{\gamma} = \sqrt{\omega_0^2 - \gamma^2 / 4}$

Critical damping is reached when $\omega_0^2 = \gamma^2 / 4$, i.e. when $\gamma = \pm 2 \omega_0$

To write this second-order ODE on standard form we introduce $p = m \frac{\mathrm{d}x}{\mathrm{d}t}$:

$\displaystyle \frac{\mathrm{d}p}{\mathrm{d}t} = - \gamma p - m \omega^2_0 x$

$\displaystyle \frac{\mathrm{d}x}{\mathrm{d}t} = \frac{p}{m}$

```python
def dy(y, t, gamma, w0, m):
    x, p = y[0], y[1]
    
    dx = p / m
    dp = - gamma * p - m * w0 * w0 * x
    
    return [dx, dp]
```

```python
# initial state
y0 = [1.0, 0.0]
```

```python
# time coodinate to solve the ODE for
t = linspace(0, 6, 1000)
w0 = 2*pi*1.0

m = 1.0
```

```python
y_undamped = odeint(dy, y0, t, args = (0.0, w0, m))
y_underd = odeint(dy, y0, t, args = (w0/4, w0, m))
y_crit = odeint(dy, y0, t, args = (2*w0, w0, m))
y_overd = odeint(dy, y0, t, args = (4*w0, w0, m))
y_test = odeint(dy, y0, t, args = (0.75*w0, w0, m))

```

```python
fig, ax = plt.subplots(figsize=(9,6))
ax.plot(t, y_undamped[:,0],'k',linewidth=0.25, label="undamped")
ax.plot(t, y_underd[:,0], label="underdamped")
ax.plot(t, y_crit[:,0], label="crit. damped")
ax.plot(t, y_overd[:,0], label="overdamped")
ax.plot(t, y_test[:,0], label="test")

ax.legend()
plt.ylabel('x')
plt.xlabel('t')
```
